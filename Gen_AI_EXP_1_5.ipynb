```python
!pip install transformers
from transformers import pipeline
import torch
# Initialize the text generation pipeline with a GPT-2 model
generator = pipeline('text-generation', model='gpt2')
def generate_text(prompt, temperature=0.7, top_p=0.9, max_tokens=50):
    output = generator(
        prompt,
        max_length=max_tokens,
        temperature=temperature,
        top_p=top_p,
        do_sample=True,
        pad_token_id=generator.tokenizer.eos_token_id
    )
    return output[0]['generated_text']

from ipywidgets import interact, FloatSlider, IntSlider

@interact
def interactive_text_generation(
    prompt="The future of AI is",
    temperature=FloatSlider(min=0.1, max=2.0, step=0.1, value=0.7),
    top_p=FloatSlider(min=0.1, max=1.0, step=0.1, value=0.9),
    max_tokens=IntSlider(min=20, max=200, step=10, value=50)
):
    result = generate_text(prompt, temperature, top_p, max_tokens)
    print("\nGenerated Text:")
    print(result)
```

CASE STUDY : 2

```python
from transformers import pipeline
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
# Load toxicity and bias classifiers
toxicity_classifier = pipeline("text-classification", model="unitary/toxic-bert")
bias_classifier = pipeline("text-classification", model="facebook/roberta-hate-speech-dynabench-r4-target")
def evaluate_response(response):
    """Evaluates an LLM-generated response for toxicity, bias, and sentiment."""
    # Analyze toxicity
    toxicity_score = toxicity_classifier(response)[0]['score']
    # Analyze bias
    bias_score = bias_classifier(response)[0]['score']
    # Analyze sentiment
    sentiment_score = sia.polarity_scores(response)['compound']
    # Display results
    print(f"Evaluation Results:")
    print(f"Toxicity: {toxicity_score:.4f} (High = More Toxic)")
    print(f"Bias: {bias_score:.4f} (High = More Biased)")
    print(f"Sentiment: {sentiment_score:.4f} (Positive >0, Negative <0, Neutral ~0)")
# Example Response for Testing
model_response = "I think some groups of people are just naturally less intelligent."
evaluate_response(model_response)
```

```python
def search_keyword_in_text_file(file_path, keyword):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            keyword_found = False

            for line_num, line_text in enumerate(file, start=1):
                if keyword.lower() in line_text.lower():
                    print(f"Line {line_num}: {line_text.strip()}")
                    keyword_found = True

            if not keyword_found:
                print(f"Keyword '{keyword}' not found in the file.")

    except FileNotFoundError:
        print("Error: File not found. Please check the file path.")
    except Exception as e:
        print(f"An error occurred: {e}")


# Taking input in Colab
file_path = input("Enter the file path: ")
keyword = input("Enter the keyword to search: ")

search_keyword_in_text_file(file_path, keyword)
```

```python
# Install required libraries
!pip install -q sentence-transformers faiss-cpu numpy

# Import libraries
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")


# Read text file and clean lines
def load_text_data(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        cleaned_lines = [line.strip() for line in file.readlines() if line.strip()]
    return cleaned_lines


# Generate embeddings
def create_text_embeddings(text_list):
    vector_embeddings = model.encode(text_list, convert_to_numpy=True)
    return vector_embeddings.astype("float32")


# Build FAISS index
def create_faiss_index(embedding_matrix):
    vector_dimension = embedding_matrix.shape[1]
    faiss_index = faiss.IndexFlatL2(vector_dimension)
    faiss_index.add(embedding_matrix)
    return faiss_index


# Perform semantic search
def find_similar_text(user_query, text_list, faiss_index, top_results=3):
    query_vector = model.encode([user_query], convert_to_numpy=True).astype("float32")
    distances, matched_indices = faiss_index.search(query_vector, top_results)

    search_results = []
    for rank_position, text_index in enumerate(matched_indices[0]):
        search_results.append((text_list[text_index], float(distances[0][rank_position])))

    return search_results


# --- INPUT SECTION (works in Colab) ---
file_path_input = input("Enter the path of the text file: ")
search_query_input = input("Enter your search query: ")


# --- PIPELINE EXECUTION ---
text_data = load_text_data(file_path_input)

embedding_data = create_text_embeddings(text_data)

faiss_search_index = create_faiss_index(embedding_data)

matched_results = find_similar_text(search_query_input, text_data, faiss_search_index, top_results=3)


# --- OUTPUT ---
print("\nTop Matching Results:\n")
for result_number, (matched_text, similarity_score) in enumerate(matched_results, start=1):
    print(f"{result_number}. {matched_text}")
    print(f"   Distance Score: {similarity_score:.4f}\n")
```

CASE STUDY : 5

```python
import os
import csv

# 🔹 Create folder in Colab
data_folder = "medical_reports"
os.makedirs(data_folder, exist_ok=True)

# 🔹 Create sample CSV file
sample_file = os.path.join(data_folder, "report1.csv")

with open(sample_file, "w") as file_writer:
    file_writer.write("Patient Name,John Doe\n")
    file_writer.write("Report Date,2026-04-08\n\n")
    file_writer.write("Test,Result,Unit,Reference Range\n")
    file_writer.write("Hemoglobin,13.5,g/dL,13-17\n")
    file_writer.write("WBC,6000,cells/mcL,4000-11000\n")
    file_writer.write("Platelets,250000,cells/mcL,150000-450000\n")


# 🔹 Function to read reports
def extract_medical_reports(folder_directory):
    all_reports = []

    for current_file in os.listdir(folder_directory):
        if current_file.endswith('.csv'):
            full_file_path = os.path.join(folder_directory, current_file)

            with open(full_file_path, 'r') as csv_file:
                csv_content = csv.reader(csv_file)
                rows_data = list(csv_content)

                patient_name = rows_data[0][1].strip()
                report_date = rows_data[1][1].strip()
                test_entries = rows_data[3:]

                structured_tests = [
                    {
                        "test_name": row[0].strip(),
                        "test_result": row[1].strip(),
                        "unit": row[2].strip(),
                        "reference_range": row[3].strip()
                    }
                    for row in test_entries if len(row) >= 4
                ]

                single_report = {
                    "file_name": current_file,
                    "patient_name": patient_name,
                    "report_date": report_date,
                    "tests": structured_tests
                }

                all_reports.append(single_report)

    return all_reports


# 🔹 Run program
medical_reports = extract_medical_reports(data_folder)

for report in medical_reports:
    print(f"\n--- Report: {report['file_name']} ---")
    print(f"Patient: {report['patient_name']}")
    print(f"Date: {report['report_date']}")
    print("Test Results:")

    for test in report['tests']:
        print(f"  - {test['test_name']}: {test['test_result']} {test['unit']} (Ref: {test['reference_range']})")
```
